# แปลงข้อมูลให้เหมาะสม


**ทำให้ชื่อคณะเป็นมาตรฐานเดียวกัน**

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'คณะวิทยาศาสตร์', 'วิทย์', 'Science', 'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['วิศวกรรมศาสตร์', 'วิศวกรรม', 'วิศวฯ', 'วิศวะ', 'Engineering']:
        return 'วิศวกรรมศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Liberal Arts', 'Arts']:
        return 'ศิลปศาสตร์'
    else:
        return name


df['คณะ'] = df['คณะ'].apply(standardize_faculty)


df['คณะ'].fillna(df['คณะ'].mode()[0], inplace=True)

```

**เปลี่ยนชื่อเพศให้เหมือนกัน**

```python
def standardize_faculty(name):
    if name in ['ชาย', 'ช', 'Male', 'M']:
        return 'ชาย'
    elif name in ['หญิง', 'ญ', 'Female', 'F']:
        return 'หญิง'
    else:
        return name

df['เพศ'] = df['เพศ'].apply(standardize_faculty)

df['เพศ'].fillna(df['เพศ'].mode()[0], inplace=True)
```

**สมมติมีคอลัมน์ 'วันที่ลงทะเบียน' เป็น string แปลงเป็น datetime**

```python
df['วันที่ลงทะเบียน'] = pd.to_datetime(df['วันที่ลงทะเบียน'])

df['ปี'] = df['วันที่ลงทะเบียน'].dt.year
df['เดือน'] = df['วันที่ลงทะเบียน'].dt.month
df['วัน'] = df['วันที่ลงทะเบียน'].dt.day
df['ชื่อวัน'] = df['วันที่ลงทะเบียน'].dt.day_name()
df
```

**ค่าต่ำสุด-สูงสุดของส่วนสูงและน้ำหนัก**

```python
max_height = df['ส่วนสูง_cm'].max()
min_height = df['ส่วนสูง_cm'].min()

print(f"ส่วนสูง_cm: สูงสุด = {max_height}, ต่ำสุด = {min_height}")


mean_cm = df['ส่วนสูง_cm'].mean()


df['ส่วนสูง_cm'].fillna(mean_cm, inplace=True)


max_weight = df['น้ำหนัก_kg'].max()
min_weight = df['น้ำหนัก_kg'].min()

print(f"น้ำหนัก_kg: สูงสุด = {max_weight}, ต่ำสุด = {min_weight}")
```

# การจัดการค่าผิดปกติ



**ใช้ IQR (Interquartile Range)**


```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]

print(f"\nขอบเขตล่าง - {lower_bound}")
print(f"ขอบเขตบน - {upper_bound}")
print(f"\nข้อมูลหลังกรองค่าผิดปกติ")
print(df_no_outliers)
print(f"\nค่าเฉลี่ยอายุ (หลังกรองค่าผิดปกติ) = {df_no_outliers['อายุ'].mean():.2f}")
```

# สร้าง features ใหม่

**สร้าง Feature BMI ถ้ามีคอลัมน์ น้ำหนัก_kg และ ส่วนสูง_cm**

```python
df['BMI'] = df['น้ำหนัก_kg'] / (df['ส่วนสูง_cm']/100)**2
df['สุขภาพ'] = pd.cut(df['BMI'],bins=[0, 18.5, 25, 30, 100],
                                labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])

print(df[['ชื่อ', 'BMI', 'สุขภาพ']])
```

```python
print(df['คณะ'].unique())
data_encoded = pd.get_dummies(df, columns=['คณะ'])
print(data_encoded)
```
